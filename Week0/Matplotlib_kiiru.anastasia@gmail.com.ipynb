#### Boxplot

```python
land_ownership = data['Q5'].tolist()
plt.boxplot(land_ownership)
plt.title("A Boxplot showing land ownership status of the population")
plt.ylabel("Land Ownership Status")
plt.show();
```

#### Barplot

```python
age = data['Q1'].tolist()
literacy_in_english = data['Q19'].tolist()
plt.bar(literacy_in_english, age)
plt.title("A Bar plot of age and level of literacy in English")
plt.xlabel("Level of Literacy in English")
plt.ylabel("Age")
plt.show;
```

#### Histogram

```python
education_level = data['Q4'].tolist()
plt.hist(education_level, bins = 7)
plt.title("A Histogram showing the distribution of highest level of education in a population")
plt.xlabel("Highest Level of Education")
plt.ylabel("Frequency")
plt.show;
```

#### Scatter Plot

```python
longitude = data['Longitude'].tolist()
latitude = data['Latitude'].tolist()
plt.scatter(longitude, latitude)
plt.title("A scatter plot showing the relationship between the longitude and the latitude of the population")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.show()
```

## Tasks 1

Import the pandas_tasks.csv

```python
import pandas as pd

import seaborn as sns
sns.set()
```

```python
data = pd.read_csv('pandas_tasks.csv')
```

#### Question 1. Make a scatter plot of the Longitude and Latitude columns

```python
sns.scatterplot(x = 'Longitude', y = 'Latitude', data = data)
```

#### Question 2. Make a scatter plot of the Longitude and Latitude columns. This time, plot with 'hue' = Q2 and compare it over Q3

```python
sns.relplot(x = 'Longitude', y = 'Latitude', hue = 'Q2', col = 'Q3', data = data)
```

#### Question 3. Make a boxplot of Longitude and Q2

```python
sns.boxplot(x = 'Q2', y = 'Longitude', data = data)
```

#### Question 4. Make a count plot of all the categorical variables in the dataset.

```python
categorical_variables = data[['Q2', 'Q3', 'Q4', 'Q5', 'Q6', 'Q7', 'Q8_1', 'Q8_2', 'Q8_3', 'Q8_4', 'Q8_5', 'Q8_6', 'Q8_7', 'Q8_8', 'Q8_9', 'Q8_10', 'Q8_11', 'Q9', 'Q10', 'Q11', 'Q12', 'Q13', 'Q14', 'Q15', 'Q16', 'Q17', 'Q18', 'Q19', 'mobile_money', 'savings', 'borrowing', 'insurance', 'mobile_money_classification']]
sns.countplot(x = 'Q2', data = categorical_variables)
```
